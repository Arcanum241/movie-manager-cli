```python

#    GLOBAL VARIABLES A
movies_list = []
file_path = 'C:/Users/User/Desktop/movies.txt' 

def load_data():
    
    """ Reads data from the text file and populates the movies_list """
    
    movies_list.clear() 
    
    try:   # exception handling 
        with open(file_path, "r", encoding = 'utf-8') as file:  #we open the file in read mode
            lines = file.readlines()

        for i in range(1, len(lines)):
            line = lines[i].strip()       #
            if line != "":
                parts = line.split(',')

                movie = {
                    "id": int(parts[0]),
                    "title": parts[1],
                    "year": int(parts[2]),
                    "genre": parts[3],
                    "rating": float(parts[4]),
                    "duration": int(parts[5])}
                movies_list.append(movie)
        print("Successful Load")

    except FileNotFoundError:
        print(f"Warning: File not found at {file_path}. Starting with empty list.")


def save_data():
    
    """ Writes our list back to the text file """
    try:
        
        with open(file_path, "w", encoding = 'utf-8') as file:
            file.write("ID,Title,Year,Genre,Rating,Duration\n")
            
            for movie in movies_list:
                line = f"{movie['id']},{movie['title']},{movie['year']},{movie['genre']},{movie['rating']},{movie['duration']}\n"
                file.write(line)
                
        print("Data saved to file successfully.")
    except Exception as e:
        print(f"Error saving data: {e}")

#2 Create methods to implement CRUD operations

def create_movie():
    print("--------Add a New Movie--------")
    try:
        MovieId = int(input("Enter Movie ID: "))
        title = input("Enter Title: ")
        year = int(input("Enter Year: "))
        genre = input("Enter Genre: ")
        rating = float(input("Enter Rating: "))
        duration = int(input("Enter Duration (min): "))

        # Create the dictionary
        new_movie = {
            "id": MovieId,
            "title": title,
            "year": year,
            "genre": genre,
            "rating": rating,
            "duration": duration
        }

        movies_list.append(new_movie)
        save_data() # Save immediately after adding
        print("Movie added.")
    except ValueError:
        print("Invalid input! Please enter numbers for ID, Year, Rating, and Duration.")

def read_movies():
    print("--------List of Movies-------")
    if not movies_list:
        print("No movies found.")
    else:
        for movie in movies_list:
            print(f"ID: {movie['id']}: {movie['title']} - {movie['year']} - {movie['genre']} - {movie['rating']}/10 - {movie['duration']} minutes")

def update_movie():
    print("-------Update Movie Rating-------")
    try:
        search_id = int(input("Enter ID to update: "))

        found = False
        for movie in movies_list:
            if movie['id'] == search_id:
                new_rating = float(input("Enter new rating: "))
                movie['rating'] = new_rating
                save_data()
                print("Rating updated.")
                found = True
                break
        
        if not found:
            print("ID was not found")
    except ValueError:
        print("Invalid input.")

def delete_movie():
    print("-------Delete a Movie-------")
    try:
        search_id = int(input("Enter ID to delete: "))

        found = False
        for movie in movies_list:
            if movie['id'] == search_id:
                movies_list.remove(movie)
                save_data()
                print("Movie deleted.")
                found = True
                break
        
        if not found:
            print("ID was not found")
    except ValueError:
        print("Invalid input.")


# 4. ALGORITHMS (Search, Sort, Filter, Max) 
# Create minimum 4 useful methods to implement the basic algorithms learned in the lectures/seminars.


def linear_search_title():       #LINEAR SEARCH ALGORITHM
    print("\n------Search Movie by Title-------")
    target = input("Enter movie title to find: ")
    
    n = len(movies_list)
    i = 0

    while i < n and movies_list[i]['title'].lower() != target.lower():
        i = i + 1
        
    exists = (i < n)
    
    if exists:
        print(f"Found! ID: {movies_list[i]['id']}, Year: {movies_list[i]['year']}, Rating: {movies_list[i]['rating']}")
    else:
        print("Movie not found.")
def maximum_rating(): # MAXIMUM SELECTION ALGORITHM
    print("\n-------Highest Rated Movie-------")
    
    if len(movies_list) == 0:
        print("List is empty.")
        return

    max = 0
    n = len(movies_list)
    
    for i in range(1, n):
        if movies_list[i]['rating'] > movies_list[max]['rating']:
            max = i
            
    best_movie = movies_list[max]
    print(f"Highest Rated: {best_movie['title']} ({best_movie['rating']})")
    
def filter_year(): #Selection (elements with P property)
    print("\n--- Filter Movies by Year ---")
    try:
        year_limit = int(input("Enter year (show movies released AFTER this year): "))
        
        results = []
        n = len(movies_list)
        
        j = -1 
        
        for i in range(n):
            if movies_list[i]['year'] > year_limit:
                j = j + 1
                results.append(movies_list[i])
                
        print(f"Found {len(results)} movies:")
        for movie in results:
            print(f"{movie['title']} ({movie['year']})")
            
    except ValueError:
        print("Invalid year.")
        
def bubble_sort_year(): #BUBBLE SORT 
    print("\n-------Sorting Movies by Year (Oldest to Newest)-------")
    
    n = len(movies_list)
    
    for i in range(n - 1, 0, -1):
        for j in range(i):
            if movies_list[j]['year'] > movies_list[j + 1]['year']:
                temp = movies_list[j]
                movies_list[j] = movies_list[j + 1]
                movies_list[j + 1] = temp
                
    print("Movies sorted by year successfully!")
    read_movies()
    
def bubble_sort_year_descending(): # BUBBLE SORT BUT REVERSED
    print("\n--- Sorting Movies by Year (Newest to Oldest) ---")
    
    n = len(movies_list)
    
    for i in range(n - 1, 0, -1):
        for j in range(i):

            if movies_list[j]['year'] < movies_list[j + 1]['year']:
                temp = movies_list[j]
                movies_list[j] = movies_list[j + 1]
                movies_list[j + 1] = temp
                
    print("Movies sorted by year (descending) successfully!")
    read_movies()


# 5. CALCULATIONS (Math/Stats)  E
#Create minimum 4 useful methods to implement mathematical, statistical, or business calculations.

def average_rating():
    """ 1. Calculates average rating """
    print("\n--- Average Movie Rating ---")
    n = len(movies_list)
    
    if n == 0:
        print("No movies to calculate.")
        return

    total_score = 0
    for movie in movies_list:
        total_score = total_score + movie['rating']
    
    avg = total_score / n
    print(f"The average rating of your library is: {avg:.2f}")

def total_duration():
    """ 2. Calculates total watch time """
    print("\n--- Total Watch Time ---")
    total_min = 0
    
    for movie in movies_list:
        total_min = total_min + movie['duration']
        
    hours = total_min // 60
    minutes = total_min % 60
    
    print(f"Total duration: {total_min} minutes")
    print(f"That is approx: {hours} hours and {minutes} minutes.")

def percentage_by_year():
    """ 3. Calculates percentage of movies from a specific year """
    print("\n--- Percentage by Year ---")
    try:
        target_year = int(input("Enter year to analyze: "))
        
        n = len(movies_list)
        if n == 0:
            print("List is empty.")
            return

        count = 0
        for movie in movies_list:
            if movie['year'] == target_year:
                count = count + 1
                
        percent = (count / n) * 100
        print(f"Movies from {target_year}: {count}")
        print(f"This represents {percent:.1f}% of your collection.")
    except ValueError:
        print("Invalid year.")

def average_duration():
    """ 4. Calculates average movie length """
    print("\n--- Average Movie Duration ---")
    n = len(movies_list)
    
    if n == 0:
        print("No movies to calculate.")
        return

    total_min = 0
    for movie in movies_list:
        total_min = total_min + movie['duration']
        
    avg_dur = total_min / n
    print(f"The average movie length is: {avg_dur:.0f} minutes.")


# 8. MAIN PROGRAM (Menu)

def main():
    print("   MOVIE MANAGEMENT SYSTEM v1.0")
    
    load_data()
    
    while True:
        print("\n--- MAIN MENU ---")
        print("1.  [CRUD] List All Movies")
        print("2.  [CRUD] Add New Movie")
        print("3.  [CRUD] Update Movie Rating")
        print("4.  [CRUD] Delete Movie")
        print("5.  [ALGORITHMS] Search Movie by Title")
        print("6.  [ALGORITHMS] Find Highest Rated Movie")
        print("7.  [ALGORITHMS] Filter Movies by Year")
        print("8.  [ALGORITHMS] Sort Movies by Year (Oldest -> Newest)")
        print("9.  [CALC] Average Rating")
        print("10. [CALC] Total Duration")
        print("11. [CALC] Percentage by Year")
        print("12. [CALC] Average Duration")
        print("13. [ALGORITHMS] Sort Movies by Year (Newest -> Oldest)") # NEW OPTION
        
        print("0.  Exit and Save")

        choice = input("\nEnter your choice: ")

        if choice == '1':
            read_movies()
        elif choice == '2':
            create_movie()
        elif choice == '3':
            update_movie()
        elif choice == '4':
            delete_movie()
        elif choice == '5':
            linear_search_title()
        elif choice == '6':
            maximum_rating()
        elif choice == '7':
            filter_year()
        elif choice == '8':
            bubble_sort_year()
        elif choice == '9':
            average_rating()
        elif choice == '10':
            total_duration()
        elif choice == '11':
            percentage_by_year()
        elif choice == '12':
            average_duration()
        elif choice == '13':
            bubble_sort_year_descending() # CALLS THE NEW FUNCTION
            
        elif choice == '0':
            print("Saving data and exiting. Goodbye!")
            save_data()
            break
        
        else:
            print("Invalid choice. Please try again.")

# Run the program
if __name__ == "__main__":
    main()
```
